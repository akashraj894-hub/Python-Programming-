# pandas

## Creating Pandas Series

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=["Rank1", "Rank2", "Rank3", "Rank4", "Rank5"]
)

print(marks)
```

## Indexing and Slicing

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=["Rank1", "Rank2", "Rank3", "Rank4", "Rank5"]
)
print(marks[0])
```

```python
print(marks.loc[["Rank1", "Rank2", "Rank3"]])
```

```python
print(marks.iloc[2])
```

```python
print(marks[marks > 90])
```

## Mnaipulating Series

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=["Rank1", "Rank2", "Rank3", "Rank4", "Rank5"]
)
marks["Rank1"] = 100
print(marks)
```

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=["Rank1", "Rank2", "Rank3", "Rank4", "Rank5"]
)
marks = marks.drop("Rank5")
print(marks)
```

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=["Rank1", "Rank2", "Rank3", "Rank4", "Rank5"]
)
cgpa = marks / 10
print(cgpa)
```

# Pandas Data Frame

## Creating Pandas Data Frame

```python
import pandas as pd

transactions = pd.DataFrame({
    "TransactionID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "ProductCategory": [
        "Electronics", "Clothing", "Electronics", "Furniture", "Clothing",
        "Electronics", "Furniture", "Clothing", "Furniture", "Electronics"
    ],
    "Region": [
        "North", "South", "North", "East", "West",
        "North", "East", "West", "South", "North"
    ],
    "Amount": [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})

print(transactions)
```

## Data Exploration

```python
print(transactions.head())
print(transactions.tail())
print(transactions.shape)
print(transactions.columns)
print(transactions.dtypes)
```

```python
print(transactions[["ProductCategory", "Amount"]])
```

```python
print(transactions.iloc[:, -3:])
```

```python
print(transactions.loc[:, ["ProductCategory", "Region", "Amount"]])
```

```python
filtered = transactions[(transactions["Region"] == "North") & (transactions["Amount"] > 200)]
print(filtered)
```

```python
print(transactions["ProductCategory"].value_counts())
```

```python
print(transactions["Region"].unique())
```

```python
print(transactions.groupby("Region")["Amount"].mean())
```

## Manipulating the Data frames

```python
transactions.loc[transactions["TransactionID"] == 102, "Amount"] = 165
print(transactions)
```

```python
transactions["Discount"] = transactions["Amount"] * 0.10
print(transactions)
```

```python
transactions = transactions[transactions["TransactionID"] != 109]
print(transactions)
```

```python
transactions = transactions.drop("Discount", axis=1)
print(transactions)
```
