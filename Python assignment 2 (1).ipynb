#List Operations / Modifications

```python
name_list =["Arun","Akash","Ravi","Rahul","Jagan"]
name_list.append('Yazhini')
print(name_list)
```

```python
age_list=[24,25,26,27,28]
age_list.insert(2,30)

print(age_list)
```

```python
name_list.remove('Yazhini')
print(name_list)
```

```python
age_list.pop(-1)
```

```python
age_list.extend([29,30,26])
age_list
```

```python
age_list.sort(reverse=True)
age_list
```

```python
max(age_list)
```

```python
min(age_list)
```

#Operators & Conditional Statements

```python
scoreofstudent= int(input('Enter score of student 0-10: '))

if scoreofstudent > 7 and scoreofstudent <= 10:
    print('Above Average - Extraordinary')
elif scoreofstudent >= 4 and scoreofstudent <= 7:
    print('Average - Good Effort')
else:
    print('Below Average - Work Hard')
```
